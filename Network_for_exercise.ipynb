```
class NetworkForExercize:
  def __init__(self):
    np.random.seed(0)
    self.w1=np.random.randn()
    self.w2=np.random.randn()
    self.w3=np.random.randn()
    self.w6=np.random.randn()
    self.w7=np.random.randn()
    self.w8=np.random.randn()
    self.w9=np.random.randn()
    self.w10=np.random.randn()
    self.w11=np.random.randn()
    self.w12=np.random.randn()
    self.b1=0
    self.b2=0
    self.b3=0
    self.b4=0
    self.b5=0
    
  def sigmoid(self,x):
    return 1/(1+np.exp(-x))
  
  # Farward pass
  def farward_pass(self,x):
    # Input layer
    self.x1,self.x2=x
    # First hidden layer
    self.a1=self.w1*self.x1+self.w2*self.x2+self.b1
    self.h1=self.sigmoid(self.a1)
    self.a2=self.w3*self.x1+self.w6*self.x2+self.b2
    self.h2=self.sigmoid(self.a2)
    # Second hidden layer
    self.a3=self.w7*self.h1+self.w8*self.h2+self.b3
    self.h3=self.sigmoid(self.a3)
    self.a4=self.w9*self.h1+self.w10*self.h2+self.b4
    self.h4=self.sigmoid(self.a4)
    # Output layer
    self.a5=self.w11*self.h3+self.w12*self.h4+self.b5
    self.h5=self.sigmoid(self.a5)
    return self.h5
  # Back Propogation step
  
  def grad(self,x,y):
    # Output layer
    self.farward_pass(x)
    self.dw11=(self.h5-y)*self.h5*(1-self.h5)*self.h3
    self.dw12=(self.h5-y)*self.h5*(1-self.h5)*self.h4
    self.db5=(self.h5-y)*self.h5*(1-self.h5)
    
    self.dw7=(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)*self.h1
    self.dw8=(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)*self.h2
    self.db3=(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)
    
    self.dw10=(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)*self.h2
    self.dw9=(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)*self.h1
    self.db4=(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)
    
    self.dw1=(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)*self.w7*self.h1*(1-self.h1)*self.x1+(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)*self.w9*self.h1*(1-self.h1)*self.x1
    self.dw2=(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)*self.w7*self.h1*(1-self.h1)*self.x2+(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)*self.w9*self.h1*(1-self.h1)*self.x2
    self.db1=(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)*self.w7*self.h1*(1-self.h1)+(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)*self.w9*self.h1*(1-self.h1)
    
    self.dw6=(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)*self.w10*self.h2*(1-self.h2)*self.x2+(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)*self.w8*self.h2*(1-self.h2)*self.x2
    self.dw3=(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)*self.w10*self.h2*(1-self.h2)*self.x1+(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)*self.w8*self.h2*(1-self.h2)*self.x1
    self.db2=(self.h5-y)*self.h5*(1-self.h5)*self.w12*self.h4*(1-self.h4)*self.w10*self.h2*(1-self.h2)+(self.h5-y)*self.h5*(1-self.h5)*self.w11*self.h3*(1-self.h3)*self.w8*self.h2*(1-self.h2)
    
  def fit(self,X,Y,epochs=1,learning_rate=1,initialise=True,display_loss=False,display_weight=False):
    
    # if initialise w,b
    if initialise:
      np.random.seed(0)
      self.w1=np.random.randn()
      self.w2=np.random.randn()
      self.w3=np.random.randn()
      self.w6=np.random.randn()
      self.w7=np.random.randn()
      self.w8=np.random.randn()
      self.w9=np.random.randn()
      self.w10=np.random.randn()
      self.w11=np.random.randn()
      self.w12=np.random.randn()
      self.b1=0
      self.b2=0
      self.b3=0
      self.b4=0
      self.b5=0
      
    if display_loss:
      loss={}
      
      
    for i in tqdm_notebook(range(epochs),total=epochs,unit='epoch'):
      dw1,dw2,dw3,dw6,dw7,dw8,dw9,dw10,dw11,dw12,db1,db2,db3,db4,db5=[0]*15
      for x,y in zip(X,Y):
        self.grad(x,y)
        
        dw1+=self.dw1
        dw2+=self.dw2
        dw3+=self.dw3
        dw6+=self.dw6
        dw7+=self.dw7
        dw8+=self.dw8
        dw9+=self.dw9
        dw10+=self.dw10
        dw11+=self.dw11
        dw12+=self.dw12
        db1+=self.db1
        db2+=self.db2
        db3+=self.db3
        db4+=self.db4
        db5+=self.db5
        
      m=X.shape[0]
      self.w1-=learning_rate*dw1/m
      self.w2-=learning_rate*dw2/m
      self.w3-=learning_rate*dw2/m
      self.w6-=learning_rate*dw6/m
      self.w7-=learning_rate*dw7/m
      self.w8-=learning_rate*dw8/m
      self.w9-=learning_rate*dw9/m
      self.w10-=learning_rate*dw10/m
      self.w11-=learning_rate*dw11/m
      self.w12-=learning_rate*dw12/m
      self.b1-=learning_rate*db1/m
      self.b2-=learning_rate*db2/m
      self.b3-=learning_rate*db3/m
      self.b4-=learning_rate*db4/m
      self.b5-=learning_rate*db5/m
      
      if display_loss:
        Y_pred=self.predict(X)
        loss[i]=mean_squared_error(Y_pred,Y)
        
      if display_weight:
        weight_matrix=np.array([[0,self.b5,self.w11,self.w12,0,0],[self.b3,self.w7,self.w8,self.w9,self.w10,self.b4],[self.b1,self.w1,self.w2,self.w3,self.w6,self.b2]])
        weight_matrices.append(weight_matrix)
      
    if display_loss:
      plt.plot(loss.values())
      plt.xlabel('Epochs')
      plt.ylabel('Mean Squared Error')
      plt.show()
        
        
      
      
  def predict(self,X):
    Y_pred=[]
    for x in X:
      y_pred=self.farward_pass(x)
      Y_pred.append(y_pred)
    return np.array(Y_pred)
  
  def predict_h1(self,X):
    Y_pred=[]
    for x in X:
      y_pred=self.farward_pass(x)
      Y_pred.append(self.h1)
    return np.array(Y_pred)
  
  def predict_h2(self,X):
    Y_pred=[]
    for x in X:
      y_pred=self.farward_pass(x)
      Y_pred.append(self.h2)
    return np.array(Y_pred)
  
  def predict_h3(self,X):
    Y_pred=[]
    for x in X:
      y_pred=self.farward_pass(x)
      Y_pred.append(self.h3)
    return np.array(Y_pred)
  
  def predict_h4(self,X):
    Y_pred=[]
    for x in X:
      y_pred=self.farward_pass(x)
      Y_pred.append(self.h4)
    return np.array(Y_pred)
  
  def predict_h5(self,X):
    Y_pred=[]
    for x in X:
      y_pred=self.farward_pass(x)
      Y_pred.append(self.h5)
    return np.array(Y_pred)
  
     
     
     
      
        
        
        
      
      
      
      
      
    
    
    
    
    
    
    
    
    
    
```

```
weight_matrices=[]
nfs=NetworkForExercize()
nfs.fit(X_train,Y_train,epochs=5000,learning_rate=0.5,display_loss=True,display_weight=True)
```

```
def make_meshgrid(x, y, h=.2):
    x_min, x_max = x.min() - 0.5, x.max() + 0.5
    y_min, y_max = y.min() - 0.5, y.max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy
  
def plot_contours(ax, predict, xx, yy, **params):
    Z = predict(np.c_[xx.ravel(), yy.ravel()])   # flatning xx values using ravel
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
  
def plot_boundary():
    xx, yy = make_meshgrid(X_train[:,0], X_train[:,1])
    predict_functions = [nfs.predict_h1, nfs.predict_h2, nfs.predict_h3,nfs.predict_h4,nfs.predict_h5]
    
    for i in range(5):

      fig, ax = plt.subplots(figsize=(10,5))

      plot_contours(ax, predict_functions[i], xx, yy,
                    cmap=my_cmap, alpha=0.2)
      ax.scatter(X_train[:,0], X_train[:,1], c=Y_train, cmap=my_cmap, alpha=0.8)
      ax.set_xlim(xx.min(), xx.max())
      ax.set_ylim(yy.min(), yy.max())
      ax.set_xlabel('X1')
      ax.set_ylabel('X2')
      ax.set_title("h"+str(i+1))
      
    return True
  
plot_boundary()
```

```
def plot_heat_map(epoch):
    fig = plt.figure(figsize=(10, 1))
    sns.heatmap(weight_matrices[epoch], annot=True, cmap=my_cmap, vmin=-3, vmax=3)
    plt.title("Epoch "+str(epoch))

    fig.canvas.draw()
    image = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
    image = image.reshape(fig.canvas.get_width_height()[::-1] + (3,))

    return image
```
